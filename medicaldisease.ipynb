```python
sns.boxplot(
    x='Disease',
    y='BMI',
    data=df
)

plt.title("BMI by Disease Status")
plt.xlabel("Disease Status")
plt.ylabel("BMI")
plt.show()
```

```python
sns.violinplot(
    x='Disease',
    y='Age',
    data=df
)

plt.title("Age distribution by Disease")
plt.xlabel("Disease Status")
plt.ylabel("Age")
plt.show()
```

```python
health = df.groupby("Disease")[[
    "BloodPressure",
    "GlucoseLevel",
    "Cholesterol",
    "HeartRate",
    "BMI"

]].mean()

health.plot(kind="bar")
plt.title("Average Health Indicators by Disease Status")
plt.xlabel("Disease Status")
plt.ylabel("Average Value")
plt.show()
```

```python
disease_count = df['Alcohol'].value_counts()

plt.pie(
    disease_count.values,
    labels=disease_count.index,
    autopct='%1.1f%%'
)

plt.title("Disease Distribution")
plt.show()
```

```python
count = df["Disease"].value_counts()
plt.pie(count,labels= count.index,autopct= '%1.1f%%')
plt.title("percentage of disease")
plt.show()
```

```python
top = df.nlargest(9,'HeartRate')

sns.barplot(
    x='Disease',
    y='BMI',
    data=top
)
plt.title("Top 10 Disease")
plt.show()
```
